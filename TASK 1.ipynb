#BAR CHART
#Top 10 countries of total population in 2022

```python
# Filter for total population data
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort the data by the '2022' column in descending order
total_population_sorted = total_population_data.sort_values(by='2022', ascending=False)

# Select the top 10 countries
total_top_ten_countries = total_population_sorted.head(10)


plt.figure(figsize=(10, 6))
plt.bar(total_top_ten_countries['Country Code'], total_top_ten_countries['2022'], color='skyblue')
plt.xlabel('Country Code')
plt.ylabel('Total Population in 2022')
plt.title('Top 10 Countries by Total Population in 2022')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code",data=total_top_ten_countries,palette="coolwarm")
plt.title("Top ten countries of total population(2022)",fontsize=10)
plt.xlabel("Total poplution",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()
```
